## 📂 Numpy 개요

#### 📌 Numpy란?
- Numpy란 Numerical Python의 약자로 다차원 배열 데이터를 효과적으로 다루는 파이썬 라이브러리
- 파이썬에서 대규모의 다차원 배열, 행렬 데이터 처리를 위한 수치 연산을 가능하게 하며 내부적으로 C언어로 구현이 되어있기 때문에 메모리를 최소한으로 유지하고 관리하면서 동시에 연산 속도도 매우 빠른 라이브러리

#### 📌 Numpy를 사용하는 이유


- 데이터를 처리할 때 데이터는 숫자들의 배열로 이루어져 저장
- 위와 같은 이미지를 처리할 때 전체 이미지는 2차원 배열 데이터로 구성되어 있으며 픽셀 하나하나가 0부터 255까지의 숫자 데이터로 처리
- 이와 같이 컴퓨터가 이해할 수 있도록 데이터를 배열로 구성 · 저장 · 가공하는 것은 데이터를 처리하기 위한 가장 기본이 되는 절차
- 이 과정에서 Numpy는 리스트와 같은 파이썬의 내장 기능에 비해 대규모 데이터로 갈수록 데이터의 저장 및 가공에 있어 뛰어난 효율성을 보장하기 때문에 Numpy를 사용

#### 📌 N차원 배열 ndarray



- ndarray란 Numpy의 N차원 배열 객체
- 왼쪽 배열은 축이 1개인 1차원 배열
- 가운데 배열은 축이 2개(axis0 : 행, axis1 : 열)인 2차원 배열
- 오른쪽 배열은 축이 3개(axis0 : 행, axis1 : 열, axis2 : 깊이)인 3차원 배열이며 이와 같이 3차원 이상의 배열을 텐서(Tensor)라고 함
